```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV
df = pd.read_csv('/content/air_quality_health_impact_data_transformed.csv')

# Mostrar las primeras filas del dataset
display(df.head())

# Información general del dataset
display(df.info())

# Resumen estadístico de los datos numéricos
display(df.describe())
```

# Histograma superpuesto:


1.  Muestra la distribución de dos variables diferentes (PM2.5 y PM10) en el mismo eje de concentración.
2. Las barras azules representan la distribución de PM2.5, mientras que las barras rojas representan PM10.
3. Se usa para visualizar la frecuencia de valores dentro de rangos específicos de concentración.
4. Permite comparar cómo se distribuyen ambas variables en el mismo espacio, ayudando a identificar patrones o diferencias en la concentración de contaminantes.


```python
# 1. Distribución de Contaminantes (PM10 y PM2.5)
plt.figure(figsize=(12, 6))
sns.histplot(df['PM2_5'], bins=30, color='blue', label='PM2.5')
sns.histplot(df['PM10'], bins=30, color='red', label='PM10')
plt.title('Distribución de PM2.5 y PM10')
plt.xlabel('Concentración')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

# Mapa de calor de matriz de correlación:

1.  Se utiliza para mostrar la correlación entre diferentes variables numéricas en un conjunto de datos.

2. Los valores van de -1 a 1:
*  1 (rojo intenso) indica una correlación positiva fuerte.
*  -1 (azul intenso) indica una correlación negativa fuerte.
*  0 (azul oscuro a neutro) indica que no hay correlación.

3. Ayuda a identificar relaciones entre variables como contaminantes (PM2.5, PM10, NO2, etc.) y factores meteorológicos (Temperatura, Humedad, Viento).



```python
# 2. Correlación entre Variables
corr_matrix = df[['PM2_5', 'PM10', 'NO2', 'SO2', 'O3', 'Temperature', 'Humidity', 'WindSpeed']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

# Gráfico de dispersión:

1. Se utiliza para visualizar la relación entre dos variables numéricas, en este caso, PM2.5 (eje X) y Casos Respiratorios (eje Y).
2. Cada punto representa un par de valores (PM2.5, Casos Respiratorios) en el conjunto de datos.
3. Permite identificar posibles correlaciones o patrones entre las variables.
En este gráfico, parece haber una alta densidad de puntos en ciertas zonas, lo que puede indicar una relación específica entre la concentración de PM2.5 y los casos de enfermedades respiratorias

```python
# 3. Impacto en la Salud (PM2.5 vs Casos Respiratorios)
plt.figure(figsize=(12, 6))
sns.scatterplot(x='PM2_5', y='RespiratoryCases', data=df, color='blue')
plt.title('Relación entre PM2.5 y Casos Respiratorios')
plt.xlabel('PM2.5')
plt.ylabel('Casos Respiratorios')
plt.show()
```

# Diagrama de caja y bigotes (boxplot):

1. Se utiliza para visualizar la distribución de datos y detectar valores atípicos.

2. Muestra la relación entre AQI (Índice de Calidad del Aire) y la Clase de Impacto en la Salud.

3. Cada caja representa la distribución de los datos en diferentes categorías de impacto en la salud.
*   La línea central en cada caja es la mediana.
*   Los bordes de la caja representan el rango intercuartílico (Q1 a Q3).
*   Los bigotes muestran el rango sin considerar los valores atípicos.
*   Los puntos fuera de los bigotes son valores atípicos (outliers).
    
4. Este gráfico permite analizar cómo varía el AQI según el impacto en la salud y si hay diferencias significativas entre categorías.


```python
# 4. Índice de Calidad del Aire (AQI) vs Impacto en la Salud
plt.figure(figsize=(12, 6))
sns.boxplot(x='HealthImpactClass', y='AQI', data=df)
plt.title('AQI por Clase de Impacto en la Salud')
plt.xlabel('Clase de Impacto en la Salud')
plt.ylabel('AQI')
plt.show()
```

```python
# 5. Discretización de AQI
df['AQI_Category'] = pd.cut(df['AQI'], bins=[0, 50, 100, 200, 300, 500], labels=['Bueno', 'Moderado', 'Dañino', 'Muy Dañino', 'Peligroso'])
plt.figure(figsize=(12, 6))
sns.countplot(x='AQI_Category', data=df, palette='viridis')
plt.title('Distribución de Categorías de AQI')
plt.xlabel('Categoría de AQI')
plt.ylabel('Frecuencia')
plt.show()
```

# Grafico Diagrama De Barras

# Graficos de disperción (scatter plot):

1. Eje X (Temperatura): Representa los valores de temperatura, que en este caso van desde -10 hasta 40.

2. Eje Y (Humedad): Representa los valores de humedad, que no están especificados en el texto pero se asume que están en el eje vertical.

3. Puntos: Cada punto en el gráfico representa una observación en el conjunto de datos, donde la posición del punto indica los valores de temperatura y humedad para esa observación.

4. El objetivo de este gráfico es explorar si existe alguna relación o patrón entre la temperatura y la humedad.

```python
# 6. Relación entre Temperatura y Humedad
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Temperature', y='Humidity', data=df, color='orange')
plt.title('Relación entre Temperatura y Humedad')
plt.xlabel('Temperatura')
plt.ylabel('Humedad')
plt.show()
```

```python
# 7. Relación entre Temperatura y O3
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Temperature', y='O3', data=df, color='green')
plt.title('Relación entre Temperatura y O3')
plt.xlabel('Temperatura')
plt.ylabel('O3')
plt.show()
```

```python
# 8. Relación entre Humedad y PM10
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Humidity', y='PM10', data=df, color='purple')
plt.title('Relación entre Humedad y PM10')
plt.xlabel('Humedad')
plt.ylabel('PM10')
plt.show()
```

```python
# 9. Gráfico de dispersión entre PM2.5 y Casos Respiratorios
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["PM2_5"], y=df["RespiratoryCases"], alpha=0.5, color="red")
plt.title("Relación entre PM2.5 y Casos Respiratorios")
plt.xlabel("Concentración de PM2.5")
plt.ylabel("Casos Respiratorios")
plt.show()
```
